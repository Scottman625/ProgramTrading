```python
import requests
import pandas as pd
import numpy as np
import datetime

df = pd.read_csv('D:\\stock\\2603_20220127191321.csv', encoding='ANSI')
for i in range(len(df)):
    
    print(df['日 期'][i])
```

```python
import sqlite3
import time
# 讀 sqlite DB
conn = sqlite3.connect('D:\\stock\\py.db')
for i in range(len(df)):
    sk_symbol = '2603'
    sk_date = df['日 期'][i]
    sk_time = '133000'
    sk_open = df['開盤價'][i]
    sk_high = df['最高價'][i]
    sk_low = df['最低價'][i]
    sk_close = df['收盤價'][i]
    sk_volume = df['成交量'][i]
    sk_5D = df['均價[5]'][i]
    sk_20D = df['均價[20]'][i]
    sk_60D = df['均價[60]'][i]
    conn.execute("INSERT INTO quote (symbol , date, time, open, high, low, close, volume, 均價_5D, 均價_20D, 均價_60D ) values (?,?,?,?,?,?,?,?,?,?,?)" , ( sk_symbol, sk_date, sk_time, sk_open,sk_high, sk_low, sk_close, sk_volume, sk_5D, sk_20D, sk_60D ) )
    conn.commit()
conn.close()
```
